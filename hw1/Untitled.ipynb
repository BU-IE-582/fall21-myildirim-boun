```python
import numpy as np
import matplotlib.pyplot as plt
```

Dataset Generation

```python
def generateData(N, D):
    # return N data points which is uniformly distributed between [-1, 1] of d dimensions
    return np.random.uniform(-1,1, [N, D])
```

Approximation of the volume of the unit hypersphere to the hypercube inscribing it

```python
def calculateApproximatedComparison(data):
    # finding number of points that are within distance 1 of the origin.
    numberOfPoints = 0
    for i in range(0, data.shape[0]):
        if np.linalg.norm(data[i]) <= 1:
            numberOfPoints = numberOfPoints + 1
    return numberOfPoints / data.shape[0]
```

Task 1.a

```python
ratios = []
dimensions = []
for d in range(1, 16):
    ratios.append(calculateApproximatedComparison(generateData(1000, d)))
    dimensions.append(d)
plt.figure('Scatter plot of D versus the fraction')
plt.xlabel('number of dimensions')
plt.ylabel('ratio of approximation of the volume of the unit hypersphere to the hypercube inscribing it')
plt.plot(dimensions, ratios)
```

```python
instances_list = []
for d in range(1, 16):
    instances = []
    for i in range(1, 101):
        instances.append(generateData(1000, d))
        
    instances_list.append(instances)
```

```python
nearest_neighbour_distances = []
min_distances_list = []
for d in range(0, 15):
    min_distances = 0
    for i in range(0, 100):
        min_distance = 10000000
        for j in range(0, 100):
            if i == j :
                continue
            try:
                if min_distance > np.linalg.norm(instances_list[d][i] - instances_list[d][j]):
                    min_distance = np.linalg.norm(instances_list[d][i] - instances_list[d][j])
            except:
                print(d)
                print(i)
                print(j)
        min_distances = min_distances + min_distance
    min_distances_list.append(min_distances/100)
```

```python

plt.figure('Scatter plot of D versus the avg nearest neighbour distance')
plt.xlabel('number of dimensions')
plt.ylabel('the avg nearest neighbour distance')
plt.plot(dimensions, min_distances_list)
```
